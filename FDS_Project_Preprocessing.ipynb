```python
import pandas as pd
import numpy as py
import seaborn as sns
import matplotlib as plt
import matplotlib.pyplot as plt
```

#Data Preprocessing

```python
#Loading Main Dataset

DiseaseData = pd.read_csv('dataset.csv')
DiseaseData
```

```python
#Loading Weight Dataset

symptom_weights = pd.read_csv('Symptom-severity.csv')
symptom_weights
```

```python
#Replace the underscores spaces in the symptom names

for col in DiseaseData.columns:

    DiseaseData[col] = DiseaseData[col].str.replace('_',' ')
DiseaseData.head()

symptom_weights['Symptom'] = symptom_weights['Symptom'].str.replace('_',' ')
```

```python
DiseaseData
```

```python
symptom_weights
```

```python
cols = DiseaseData.columns
data = DiseaseData[cols].values.flatten()

strip = pd.Series(data)
strip = strip.str.strip()
strip = strip.values.reshape(DiseaseData.shape)

DiseaseData = pd.DataFrame(strip, columns=DiseaseData.columns)
DiseaseData.head()
```

```python
#Imputing the NaN values as 0s

DiseaseData = DiseaseData.fillna(0)
DiseaseData.head()
```

```python
unique_values = symptom_weights['Symptom'].unique()
print(len(unique_values))
```

```python
#replace the symptoms with weights

vals = DiseaseData.values
symptoms = symptom_weights['Symptom'].unique()

for i in range(len(symptoms)):
    vals[vals == symptoms[i]] = symptom_weights[symptom_weights['Symptom'] == symptoms[i]]['weight'].values[0]

DiseaseData = pd.DataFrame(vals, columns=cols)

#replace the symptoms with no assigned weights to zero
DiseaseData = DiseaseData.replace('foul smell of urine',0)
DiseaseData = DiseaseData.replace('dischromic  patches', 0)
DiseaseData = DiseaseData.replace('spotting  urination',0)

#scramble the dataset
DiseaseData = DiseaseData.sample(frac=1).reset_index(drop=True)
DiseaseData.head(1000)
```

```python
first_column = DiseaseData.iloc[:, 0]
WeightedData = DiseaseData.drop(DiseaseData.columns[0], axis=1)
WeightedData['Disease'] = first_column

WeightedData.head(400)
```

```python
WeightedData['Disease'].unique()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Weighted_Dataset.csv')

df = df.drop('Unnamed: 0', axis=1, errors='ignore')

# Select only the symptom columns
symptom_columns = [col for col in df.columns if 'Symptom_' in col]
df_symptoms = df[symptom_columns + ['Disease']]

# Create a pivot table for the heatmap
heatmap_data = pd.pivot_table(df_symptoms, values=symptom_columns, index='Disease', aggfunc='mean', fill_value=0)

# Plot the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(heatmap_data, cmap='viridis', annot=True, fmt='.2f')
plt.title('Symptom-Disease Relationships')
plt.xlabel('Symptom')
plt.ylabel('Disease')
plt.show()
```
